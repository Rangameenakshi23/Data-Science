# Task-2:Prediction using Desicion Tree Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
from sklearn import metrics  
import seaborn as sns  
from sklearn.datasets import load_iris  
from sklearn.model_selection import train_test_split  
from sklearn import tree  
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()  
```

```python
data = pd.DataFrame(data = iris.data, columns = iris.feature_names) 
```

```python
data['Species'] = iris.target  
```

```python
target = np.unique(iris.target)  
target_n = np.unique(iris.target_names)  
target_dict = dict(zip(target, target_n))  
data['Species'] = data['Species'].replace(target_dict)  
```

```python
x = data.drop(columns = "Species")  
y = data["Species"]  
names_features = x.columns  
target_labels = y.unique()  
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 10)  
```

```python
from sklearn.tree import DecisionTreeClassifier  
```

```python
dt=DecisionTreeClassifier(max_depth=3, random_state=10)  
```

```python
dt.fit(x_train, y_train)  
```

```python
plt.figure(figsize = (30, 10), facecolor = 'c')  
Tree = tree.plot_tree(dt, feature_names = names_features, class_names = target_labels, rounded = True, filled = True, fontsize = 14)  
plt.show()  
```

```python
y_pred = dt.predict(x_test)  
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

In this project we have predicted the right class of the species based on the attributes of flower accordingly,with a accuracy of 97%.
